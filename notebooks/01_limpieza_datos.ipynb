```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cargar_datos(ruta):
    return pd.read_excel(ruta)


def preparar_datos(df):
    df["Fecha_Hora"] = pd.to_datetime(df["Fecha_Hora"])
    return df


def revisar_datos_faltantes(df):
    return df.isnull().sum()


def resumen_general(df):
    return df.describe()


ruta = "../data/datos_compost.xlsx"

df = cargar_datos(ruta)
df = preparar_datos(df)

df.head()
```

```python
revisar_datos_faltantes(df)
```

```python
resumen_general(df)
```

```python
columnas_numericas = [
    "Temp_Invernadero",
    "Humedad_Invernadero",
    "Temp_Exterior",
    "Humedad_Exterior",
    "Temp_Compost",
    "Humedad_Compost"
]

for columna in columnas_numericas:
    df[columna] = pd.to_numeric(df[columna], errors="coerce")

df.info()
```

```python
df[["Temp_Compost", "Humedad_Compost"]].describe()
```
