## 2. Importación de librerías


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## 3. Carga de datos

Aquí se carga el archivo ubicado en la carpeta `data`.


```python
df = pd.read_excel(r"C:\Users\maria\proyecto-2026\data\datos_compost.xlsx")
df_persona2 = df.iloc[495:989]
print("Forma del dataset:", df_persona2.shape)
df_persona2.head()
```

## 4. Análisis de datos

Aquí se realiza el análisis con Pandas.


```python
temp_promedio_compost = df_persona2['Temp_Compost'].mean()
temp_max_compost = df_persona2['Temp_Compost'].max()
temp_min_compost = df_persona2['Temp_Compost'].min()
```

```python
temp_promedio_inv = df_persona2['Temp_Invernadero'].mean()
temp_max_inv = df_persona2['Temp_Invernadero'].max()
temp_min_inv = df_persona2['Temp_Invernadero'].min()
```

```python
temp_promedio_ext = df_persona2['Temp_Exterior'].mean()
temp_max_ext = df_persona2['Temp_Exterior'].max()
temp_min_ext = df_persona2['Temp_Exterior'].min()
```

```python
print("=== TEMPERATURA COMPOST ===")
print(f"Promedio: {temp_promedio_compost:.2f}°C")
print(f"Máxima:   {temp_max_compost:.2f}°C")
print(f"Mínima:   {temp_min_compost:.2f}°C")

```

```python
print("\n=== TEMPERATURA INVERNADERO ===")
print(f"Promedio: {temp_promedio_inv:.2f}°C")
print(f"Máxima:   {temp_max_inv:.2f}°C")
print(f"Mínima:   {temp_min_inv:.2f}°C")
```

```python

print("\n=== TEMPERATURA EXTERIOR ===")
print(f"Promedio: {temp_promedio_ext:.2f}°C")
print(f"Máxima:   {temp_max_ext:.2f}°C")
print(f"Mínima:   {temp_min_ext:.2f}°C")
```

## 5. Gráficas

Aquí se pueden generar gráficas de los resultados.


```python
plt.figure(figsize=(14,5))
plt.plot(df_persona2['Fecha_Hora'], df_persona2['Temp_Compost'], color='orange', label='Compost')
plt.plot(df_persona2['Fecha_Hora'], df_persona2['Temp_Invernadero'], color='green', label='Invernadero')
plt.plot(df_persona2['Fecha_Hora'], df_persona2['Temp_Exterior'], color='blue', label='Exterior')
plt.axhline(y=45, color='red', linestyle='--', label='Mín óptimo (45°C)')
plt.axhline(y=65, color='red', linestyle='--', label='Máx óptimo (65°C)')
plt.title('Análisis de Temperatura - Persona 2')
plt.xlabel('Fecha y Hora')
plt.ylabel('Temperatura (°C)')
plt.legend()
plt.xticks(df_persona2['Fecha_Hora'][::50], rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
